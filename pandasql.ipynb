```python
from pandasql import sqldf
import pandas as pd
```

```python
airbnb = pd.read_csv("train_users_2.csv")
global airbnb
```

```python
# overview of the data
pysqldf = lambda q: sqldf(q, globals())
q = """
SELECT * FROM airbnb
LIMIT 5
"""
pysqldf(q)
```

```python
# check if there are duplicates in id
duplicate = '''
SELECT * FROM airbnb
GROUP BY id
HAVING COUNT(*) > 1
'''
pysqldf(duplicate)
```

```python
# check if there is any missing value in data_first_booking
missing = '''
SELECT * FROM airbnb
WHERE date_first_booking IS NULL 
'''
pysqldf(missing)
```

```python
# check if there is any missing value in data_first_booking
missing = '''
SELECT * FROM airbnb
WHERE age is NULL
'''
pysqldf(missing)
```

```python
# check if there is outlier in age
age_range = '''
SELECT MIN(age), MAX(age) FROM airbnb
'''
pysqldf(age_range)
```

```python
# check if there is outlier in age
age_outlier = '''
SELECT age FROM airbnb
WHERE age > 80
ORDER BY age DESC
'''
pysqldf(age_outlier)
```
